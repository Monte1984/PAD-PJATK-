# **ZADANIE 1**

```python
# ZADANIE 1

data = pd.read_csv('president_heights.csv')
heights = np.array(data['height(cm)']) 
print(heights) 
print()

# Średni wzrost
print("Mean height:         ", np.mean(heights))

# Odchylenie standardowe
print("Standard deviation:  ", np.std(heights))

# Najwyższy
print("Maximum height:      ", np.amax(heights))

# Najniższy
print("Minimum height:      ", np.amin(heights))

# 25 kwantyl
print("25th percentile:     ", np.percentile(heights, 25))

# Mediana
print("Median:              ", np.median(heights))

# 75 kwantyl
print("75th percentile:     ", np.percentile(heights, 75))
```

# **ZADANIE 3**

```python
# ZADANIE 3

rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values
inches = rainfall / 254.0  # 1/10mm -> inches
inches.shape
```

```python
# Dni, kiedy nie padało
without_rain = np.count_nonzero(inches == 0)
print("Number days without rain:        ", without_rain)

# Dni, kiedy padało
with_rain = np.count_nonzero(inches)
print("Number days with rain:           ", with_rain)

# Dni, kiedy spadło powyżej 0,5" deszczu
above_05 = np.count_nonzero(inches > 0.5)
print("Days with more than 0.5 inches:  ", above_05)

# Dni, kiedy spadło poniżej 0,2" deszczu, ale padało
below_02 = np.count_nonzero((inches != 0) & (inches < 0.2))
print("Rainy days with < 0.2 inches:    ", below_02)
```

```python
# Mediana opadów w deszczowe dni w 2014 roku
mask1 = np.ma.masked_equal(inches, 0)
print("Median precip on rainy days in 2014 (inches):    ", np.round(np.ma.median(mask1), 3))

# Mediana opadów latem w 2014 roku (czyli dni pomiędzy dniem 172 a 262)
mask2 = np.ma.masked_equal(inches[171:262], 0)
print("Median precip on summer days in 2014 (inches):   ", np.round(np.ma.median(mask2), 3))

# Maksymalne opady latem 2014 roku 
max_precip_summer = mask2.max()
# analogiczne do: np.max(inches[171:262])
print("Maximum precip on summer days in 2014 (inches):  ", np.round(max_precip_summer, 3))

# Maksymalne opady poza latem 2014 roku (czyli wiosna, jesień i zima)
spring_precip = inches[:171]
winter_autumn_precip = inches[263:]
precip_included = np.concatenate([spring_precip,winter_autumn_precip])
mask3 = np.ma.masked_equal(precip_included, 0)
max_precip_non_summer = mask3.max()
print("Median precip on non-summer rainy days (inches): ", np.round(max_precip_non_summer, 3))
```
